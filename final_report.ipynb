#### Feature Engineering

We have decided to engineer three features: 

1. `Year_Birth` -> `Age`
This aligns with inutitive interpretaitons regarding age (higher = older, rather than higher = younger). 
Note: the dataset was uplaoded in 2021, and has not been updated since. Thus, we will use 2021 to calculate ages.

2. `Dt_Customer` -> `Enrollment_Recency_Days`
We will use the most recent customer enrollment date to compute the number of days a customer has been enrolled. This avoids issues with models treating the datetime variable as continuous.

3. `AcceptedCmp1-5` -> `AcceptedCmpAny`
To make sure our analysis is not significantly affected by class imbalance, we will combine all the campaign acceptance variables, with 1 representing that the customer has accepted any of the 5 campaigns.

We will drop the original variables with which we engineered the new variables to avoid multicollinearity. 

```r
marketing_campaign <- marketing_campaign %>% 
    mutate(
        Age = 2021 - Year_Birth,
        Enrollment_Recency_Days = as.numeric(max(marketing_campaign$Dt_Customer) - Dt_Customer),
        AcceptedCmpAny = if_else(
            (AcceptedCmp1 == 1 | 
            AcceptedCmp2 == 1 | 
            AcceptedCmp3 == 1 |
            AcceptedCmp4 == 1 |
            AcceptedCmp5 == 1),
            1,
            0)
        ) %>%
    select(
        -Year_Birth, 
        -Dt_Customer, 
        -AcceptedCmp1, 
        -AcceptedCmp2, 
        -AcceptedCmp3, 
        -AcceptedCmp4,
        -AcceptedCmp5)

    
```

#### Feature Engineering (VERSION 2)

4. `Income` -> `Income_thousands `
We will divide `Income` by 1000 to represent Income in thousands of dollars. This is more easily interpreted than Income in singular dollars.

5. `MntSpent` -> `MntSpent_hundreds `
We will divide the amount spent variables by 100 to represent hundreds of dollars spent. This is more easily interpreted than singular dollars spent.

```r
marketing_campaign_FE_ver2 <- marketing_campaign %>%
    mutate(
        Income_thousands = Income / 1000, 
        MntWines_hundreds = MntWines / 100,
        MntFruits_hundreds = MntFruits / 100,
        MntMeatProducts_hundreds = MntMeatProducts / 100,
        MntFishProducts_hundreds = MntFishProducts / 100,
        MntSweetProducts_hundreds = MntSweetProducts / 100,
        MntGoldProds_hundreds = MntGoldProds / 100,
    ) %>%
    select(
        -Income,
        -MntWines,
        -MntFruits,
        -MntMeatProducts,
        -MntFishProducts,
        -MntSweetProducts,
        -MntGoldProds
    )
```

### Visualizations


We identified two types of variables--behavioural and dmeographic. 

##### Behavioural

This visualization plots monthly spending against monthly catalogue purchases, faceted by whether the customer has ever accepted a campaign. We have calculated the average spending per month as well as the average number of catalogue purchases per month using our engineeered Enrollment_Recency_Days variable. With these averages, we used one boxplot per spending group. We decided to calculate average spending amounts and average catalogue purchases to standardize them and avoid the effects of customer history. Additionally, we needed to account for the face that the amount spent variables only include the previous two years, while some customers had longer enrollment histories.

This visualization allows us to compare the associations between spending amount, responses to campaigns, and catalogue purchases. We can easily see how catalogue purchases vary across spending groups, and depending on whether the customer has ever accepted a campaign. Using boxplots also allows us to focus on the median values, while staying aware of the typical spread and any outliers. In the context of our question, this plot helps us identify general trends associated with certain behaviours without ignoring variability.

```r
marketing_campaign_plotting_behavioural <- marketing_campaign %>%
    mutate( #engineering features for plot
        months_enrolled =  pmax(1, floor(Enrollment_Recency_Days / 30)),
        avg_total_spending_month = (MntWines + MntFruits + MntMeatProducts + 
                                    MntFishProducts + MntSweetProducts + MntGoldProds)/pmin(months_enrolled, 24), #make sure that we are only accounting for the past 2 years
        avg_catalog_purchases_month = NumCatalogPurchases/months_enrolled,
        spending_group = cut(avg_total_spending_month, 
                            breaks = c(seq(0, 2000, by = 200), Inf),
                            include.lowest = TRUE,
                            right = TRUE,
                            labels = c("$0-200", "$200-400", "$400-600", "$600-800", "$800-1000",
                                       "$1000-1200", "$1200-1400", "$1400-1600", "$1600-1800",
                                       "$1800-2000", "$2000+")
                              ),
        AcceptedCmpAny = factor(AcceptedCmpAny,
                                levels = c(0, 1),
                                labels = c("Did Not Accept Any Campaign", "Accepted At Least One Campaign"))
    ) 


options(repr.plot.width=12, repr.plot.height=12)
ggplot(marketing_campaign_plotting_behavioural, aes(x = spending_group, y = avg_catalog_purchases_month)) +
  geom_boxplot(fill = "lightblue", color = "black") +
  facet_wrap(~ AcceptedCmpAny) +
  labs(
    title = "Catalogue Purchases by Spending Amount",
    x = "Average Monthly Spending, Across All Categories",
    y = "Number of Catalogue Purchases per Month"
  ) +
  scale_fill_brewer(palette="BuPu") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1), plot.title = element_text(hjust = 0.5, size = 14)) 
```

##### Demographics

The next visualization compares average purchases from the three different channels (store, web, catalogue). Age and Income was grouped to enable a bar chart, and averages were taken for each demographic group and purchasing channel.  The chart is faceted by demographic variable.

This allows for easy comparison between the averages of the same purchase channel between demographic groups, as well as easy comparison of the different purchasing channels within the same group. A stack bar chart, such as this plot, allows for the viewer to quickly find information such as the most popular purchase channel for each group, as well as compare the relative popularities within a group, and see the distribution across demographic variables. Showing average figures is useful in this type of context--analysing customer behaviour for the purposes of improving business operations. In the context of business operations, we imagine that aggregate insights are preferred for due to practicality, even if it meant neglecting distributions and outliers. 
 

```r
#wrangling to find at averages per group
marketing_campaign_plotting <- marketing_campaign %>%
    mutate( #creating groups with reasonable/commonly accepted boundaries 
           Age_Group = case_when(
               Age < 30 ~ 'Under 30',
               Age >= 30 & Age < 40 ~ '30 to 39',
               Age >= 40 & Age < 50 ~ '40 to 49',
               Age >= 50 & Age < 65 ~ '50 to 64',
               Age >= 65 ~ 'Over 65'),
           Income_Group = case_when(
               Income < 45000 ~ 'Under $45,000',
               Income >= 45000 & Income < 100000 ~ '45,000 to 100,000',
               Income >= 100000 & Income < 150000 ~ '100,000 to 150,000',
               Income >= 150000 ~ 'Over 150,000')
           ) %>%
    select(Age_Group, Income_Group, Education, Marital_Status, NumStorePurchases, NumCatalogPurchases, NumWebPurchases
          ) %>%
    pivot_longer(
        cols = c(Age_Group, Income_Group, Education, Marital_Status),
        names_to = 'Demographic',
        values_to = 'Group'
    ) %>%
    group_by(Demographic, Group) %>%
    summarize(
        avg_cat_purchases = mean(NumCatalogPurchases),
        avg_store_purchases = mean(NumStorePurchases),
        avg_web_purchases = mean(NumWebPurchases),
        .groups = "drop_last"
    )%>%
    pivot_longer(
        cols = c(avg_store_purchases, avg_cat_purchases, avg_web_purchases),
        names_to = 'Variable',
        values_to = 'Average'
    ) %>% 
    mutate(
        Variable = dplyr::recode(
            Variable,
            avg_cat_purchases = 'Catalogue Purchases',
            avg_store_purchases = 'Store Purchases',
            avg_web_purchases = 'Website Purchases'),
        Demographic = dplyr::recode(
            Demographic, 
            Age_Group = 'Age Group',
            Income_Group = 'Income Group',
            Marital_Status = 'Marital Status'),
        Group = factor(
            Group,
            levels = c(
                'Under 30', '30 to 39', '40 to 49', '50 to 64', 'Over 65',
                'Under $45,000', '45,000 to 100,000', '100,000 to 150,000', 'Over 150,000',
                'Basic', 'Graduation', '2n Cycle', 'Master', 'PhD', #2n Cycle is similar to a master's degree
                'Single', 'Together', 'Married', 'Divorced', 'Widow'
                ),
            ordered = TRUE
        )
    ) 

options(repr.plot.width=12, repr.plot.height=24)
marketing_campaign_plotting %>%
    ggplot(aes(x = Group, y = Average, fill = Variable)) +
    geom_bar(position = 'dodge', stat = 'identity') + 
    facet_wrap(~ Demographic, scales = 'free', ncol = 1) + 
    theme(
        axis.text.x = element_text(angle = 45, hjust = 1),
        text = element_text(size = 20)) +
    labs(
        y = 'Average Count',
        x = 'Group Within Demographic' ,
        fill = 'Channel') +
    ggtitle('Purchases Through Different Channels By Demographic Groups')
```
