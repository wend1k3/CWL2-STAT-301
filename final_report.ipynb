#### Feature Engineering

We have decided to engineer three features: 

1. `Year_Birth` -> `Age`
This aligns with inutitive interpretaitons regarding age (higher = older, rather than higher = younger). 
Note: the dataset was uplaoded in 2021, and has not been updated since. Thus, we will use 2021 to calculate ages.

2. `Dt_Customer` -> `Recency`
We will use the most recent customer enrollment date to compute the amount of time a customer has been enrolled. This avoids issues with models treating the datetime variable as continuous.

3. `AcceptedCmp1-5` -> `AcceptedCmpAny`
To make sure our analysis is not significantly affected by class imbalance, we will combine all the campaign acceptance variables, with 1 representing that the customer has accepted any of the 5 campaigns.

We will drop the original variables with which we engineered the new variables to avoid multicollinearity. 

```r
marketing_campaign <- marketing_campaign %>% 
    mutate(
        Age = 2021 - Year_Birth,
        Recency = as.numeric(max(marketing_campaign$Dt_Customer) - Dt_Customer),
        AcceptedCmpAny = if_else(
            (AcceptedCmp1 == 1 | 
            AcceptedCmp2 == 1 | 
            AcceptedCmp3 == 1 |
            AcceptedCmp4 == 1 |
            AcceptedCmp5 == 1),
            1,
            0)
        ) %>%
    select(
        -Year_Birth, 
        -Dt_Customer, 
        -AcceptedCmp1, 
        -AcceptedCmp2, 
        -AcceptedCmp3, 
        -AcceptedCmp4,
        -AcceptedCmp5)

    
```
